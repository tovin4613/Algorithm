```python
import sys
from collections import deque

def bfs_fire(queue) :
    
    while queue :
        x, y = queue.popleft()
        for i in range(4) :
            nx = x + dx[i]
            ny = y + dy[i]
            
            if (0 <= nx < R) and (0 <= ny < C) :
                if (visited_fire[nx][ny] == 0) :
                    queue.append((nx, ny))
                    visited_fire[nx][ny] += visited_fire[x][y] + 1

def bfs_jihun(queue) :
    
    while queue :
        x, y = queue.popleft()
        for i in range(4) :
            nx = x + dx[i]
            ny = y + dy[i]
            
            if (nx < 0) or (nx >= R) or (ny < 0) or (ny >= C) :
                return visited_jihun[x][y]

            if (0 <= nx < R) and (0 <= ny < C) :
                if (visited_jihun[nx][ny] == 0) :
                    visited_jihun[nx][ny] += visited_jihun[x][y] + 1
                    if (visited_jihun[nx][ny] <= visited_fire[nx][ny]) :
                        queue.append((nx, ny))
    return 'IMPOSSIBLE'

R, C = map(int, input().split())
graph = [list(input()) for _ in range(R)]
visited_fire = [[0 for _ in range(C)] for _ in range(R)]
visited_jihun = [[0 for _ in range(C)] for _ in range(R)]
dx = [0, 0, 1, -1]
dy = [1, -1, 0, 0]

# 벽은 1로 처리
# 불 난 공간의 좌표 queue에 넣기
# 지훈이 좌표 미리 가져오기
queue = deque()
a, b = 0, 0
fire_count = 0
for i in range(R) :
    for j in range(C) :
        if graph[i][j] == '#' :
            visited_fire[i][j] = 1
            visited_jihun[i][j] = 1
        if graph[i][j] == 'F' :
            queue.append((i, j))
            visited_fire[i][j] = 1
            fire_count += 1
        if graph[i][j] == 'J' :
            a, b = i, j
            visited_jihun[i][j] = 1

bfs_fire(queue)

queue.append((a, b))

if fire_count == 0 :
    print(min(R, C))
else :
    print(bfs_jihun(queue))
```
